```python
import networkx as nx
import matplotlib.pyplot as plt


def decomposer(domain: Domain, relationship: RELATIONSHIPS):
    composition_list = []
    print("rel: ", relationship.plural)
    #target = domain.dict(include={relationship.plural})
    target = domain.get_element_container(relationship)
    print("Target: ", target)
    for c in target:
        print("c:", c.of)
        composition_list.append((c.of.name, c.by.name),)
    return composition_list
               
def decompose_compositions(domain: Domain):
    return decomposer(domain = domain, relationship = RELATIONSHIPS.COMPOSITION)

def decompose_operations(domain: Domain):
    return decomposer(domain = domain, relationship = RELATIONSHIPS.OPERATION)

def decompose_provenances(domain: Domain):
    return decomposer(domain = domain, relationship = RELATIONSHIPS.PROVENANCE)

def decompose_influences(domain: Domain):
    return decomposer(domain = domain, relationship = RELATIONSHIPS.INFLUENCE)

edgelist=[(0, 1), (1, 2), (2, 3), (3, 0)],

compositions = decompose_compositions(d26)

print("compositions: ", compositions)

#def decompose_influences(domain: Domain):
#    influence_list = []
#    for i in domain.influences:
#        influence_list.append([i.containing.of.name, i.containing.by.name])
#    for p in domain.provenances:
#        influence_list.append([p.of.name, p.by.name])
#
#    return influence_list
#
#influences = decompose_influences(d26)
#
#print("influences: ", influences)

def how_get(domain: Domain):
    influence_list = []
    operations = decompose_operations(domain = domain)
    provenances = decompose_provenances(domain = domain)
    influences = decompose_influences(domain = domain)
    graph_list = {"operations": operations, "provenances": provenances, "influences": influences}
    return graph_list
                                                            

how = how_get(d29)

pprint(how)
```

```python
G = nx.DiGraph(compositions)

print("G: ", G.nodes)

for layer, nodes in enumerate(nx.topological_generations(G)):
    # `multipartite_layout` expects the layer as a node attribute, so add the
    # numeric layer value as a node attribute
    for node in nodes:
        G.nodes[node]["layer"] = layer
        print(node, layer)

# Compute the multipartite_layout using the "layer" node attribute
pos = nx.multipartite_layout(G, subset_key="layer")
options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}


fig, ax = plt.subplots()
nx.draw_networkx(G, pos=pos, ax=ax)
print("nx: ", nx.__dict__['edgelist'])
nx.draw_networkx_edges(G, pos=pos, ax=ax, edge_color="red", width=2)
print("G 2", G.nodes)
print("pos: ", pos)

nx.draw_networkx_nodes(G, pos=pos, ax=ax, node_color="tab:red")
ax.set_title("DAG Composition in Topological Order")


plt.show()
```

```python
def coordinate_generator(graph_dict):
    raw_list = []
    for key, value in graph_dict.items():
        for item in value:
            raw_list.append(item)
    return raw_list

print(coordinate_generator(how))


G = nx.DiGraph(coordinate_generator(how))

for layer, nodes in enumerate(nx.topological_generations(G)):
    # `multipartite_layout` expects the layer as a node attribute, so add the
    # numeric layer value as a node attribute
    for node in nodes:
        G.nodes[node]["layer"] = layer
        print(node, layer)

# Compute the multipartite_layout using the "layer" node attribute
pos = nx.multipartite_layout(G, subset_key="layer")
    
options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}


fig, ax = plt.subplots()
#nx.draw_networkx(G, pos=pos, ax=ax)
print(G.nodes)
nx.draw_networkx(G, pos=pos, ax=ax)

nx.draw_networkx_nodes(G, pos=pos, ax=ax, node_color="tab:red")

nx.draw_networkx_edges(G, pos=pos, edgelist=how['operations'], ax=ax, edge_color="blue", width=2)
nx.draw_networkx_edges(G, pos=pos, edgelist=how['provenances'], ax=ax, edge_color="tab:red", width=2)
nx.draw_networkx_edges(G, pos=pos, edgelist=how['influences'], ax=ax, edge_color="tab:blue", width=2)

ax.set_title("DAG System Influence in Topological Order")


plt.show()
```
